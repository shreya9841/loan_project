```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import warnings
warnings.filterwarnings('ignore')
```

```python
dataset=pd.read_csv('C://Users//user//Desktop//7sem//loan-prediction.csv')
```

```python
print("original data")
dataset.head(5)
```

```python
del dataset["Loan_ID"]
```

```python
print("dataset after removing Loan_ID column ")
dataset
```

```python
#check missing or Null Values in columns
print("Result for checking for null  data")
dataset.isna().sum()
```

```python
# Fill missing LoanAmount values with median
dataset['LoanAmount'].fillna(dataset['LoanAmount'].median(), inplace=True)

# Fill missing Loan_Amount_Term values with median
dataset['Loan_Amount_Term'].fillna(dataset['Loan_Amount_Term'].median(), inplace=True)

# Fill missing Credit_History values with median
dataset['Credit_History'].fillna(dataset['Credit_History'].median(), inplace=True)
```

```python
# Fill missing Gender values with most frequent (mode)
dataset['Gender'].fillna(dataset['Gender'].mode()[0], inplace=True)

# Fill missing Married values with most frequent (mode)
dataset['Married'].fillna(dataset['Married'].mode()[0], inplace=True)

# Fill missing Dependents values with most frequent (mode)
dataset['Dependents'].fillna(dataset['Dependents'].mode()[0], inplace=True)

# Fill missing Self_Employed values with most frequent (mode)
dataset['Self_Employed'].fillna(dataset['Self_Employed'].mode()[0], inplace=True)
```

```python
print("After Replacing median and mode value for null data ")
dataset.isna().sum()
```

```python
# Convert Gender: Male = 0, Female = 1
dataset['Gender'] = dataset['Gender'].map({'Male': 0, 'Female': 1})

# Convert Married: No = 0, Yes = 1
dataset['Married'] = dataset['Married'].map({'No': 0, 'Yes': 1})

# Convert Education: Not Graduate = 0, Graduate = 1
dataset['Education'] = dataset['Education'].map({'Not Graduate': 0, 'Graduate': 1})

# Convert Self_Employed: No = 0, Yes = 1
dataset['Self_Employed'] = dataset['Self_Employed'].map({'No': 0, 'Yes': 1})

# Convert Property_Area using LabelEncoder
le = LabelEncoder()
dataset['Property_Area'] = le.fit_transform(dataset['Property_Area'])

# Convert Loan_Status: N = 0, Y = 1
dataset['Loan_Status'] = dataset['Loan_Status'].map({'N': 0, 'Y': 1})

# Convert Dependents: '3+' to 3, then to int
dataset['Dependents'] = dataset['Dependents'].replace('3+', 3).astype(int)
```

```python
print("data after label encoding")
dataset.head(5)
```

```python
# Save the final cleaned dataset to CSV
dataset.to_csv("cleaned_loan_data.csv", index=False)

print("Dataset saved as cleaned_loan_data.csv")
```
